```python
import os
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
from tqdm import tqdm
import pickle
import logging
logger = logging.getLogger(__name__)
logging.basicConfig(format='%(asctime)s - %(levelname)s - %(name)s -   %(message)s',
                    datefmt='%m/%d/%Y %H:%M:%S',
                    level=logging.INFO)
```

```python
with open("all_data.txt",'r') as f:
    sentences=[]
    for idx,l in enumerate(f.readlines()):
        sentences.append(l.split())
        
        
```

```python
print(len(sentences))
```

```python
sentences[0]
```

```python
w2v=Word2Vec(sentences,size=300, window=15,min_count=1,sg=1,workers=32,iter=100)
```

```python
w2v.wv.save_word2vec_format("w2v_emb.txt",binary=False)
```
