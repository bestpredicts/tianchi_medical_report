```python
import pandas as pd
```

```python
s1 = pd.read_csv("n_gram_mask_bert_v1_weight_v2/submit_0.96034970703125.csv",header=None)
```

```python
s2 = pd.read_csv("new_bert_weight_v2/submit_0.960414453125.csv",header=None)
```

```python
s1.head()
```

```python
s1[1]=s1[1].apply(lambda x:x[1:].split())
s1.head()
```

```python
s2[1]=s2[1].apply(lambda x:x[1:].split())
s2.head()
```

```python
import numpy as np

def blend(s1,s2):
    s1 = [float(i) for i in s1]
    s2 = [float(i) for i in s2]
    m = (np.array(s1)+np.array(s2))/2
    return m

s = s1.copy()
```

```python
s[2] = s2[1]
```

```python
s.head()
```

```python
s['result'] =  s.apply(lambda x:blend(x[1],x[2]),axis=1)
```

```python
s.head()
```

```python
s['submit'] =s['result'].apply(lambda x:[str(i) for i in x])
```

```python
s['submit'] = s['submit'].apply(lambda x:" ".join(x))
```

```python
s.head()
```

```python
s[0]=s[0].apply(lambda x:x[:-1])
s.head()
```

```python
sub_id=s[0].values
pres_all = s['submit'].tolist()
str_w=''
with open(f'blend.csv','w') as f:
    for i in range(len(sub_id)):
        prob= pres_all[i]
#         pres_fold=[str(p) for p in pres_fold]
#         prob = " ".join(pres_fold)
        str_w+=sub_id[i]+'|'+','+'|'+prob+'\n'
    str_w=str_w.strip('\n')
    f.write(str_w)
```
